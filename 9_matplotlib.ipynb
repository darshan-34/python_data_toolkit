```python
# bar plot >> to see the districution of categorical variables
x = ["a", "b", "c", "d"] # students
y = [4,7,1,9] # no. of days present in class

plt.bar(x,y,color='r') #vertical plot
plt.show()

plt.barh(x,y,color='r') #horizontal plot
plt.show()
```

```python
# pie chart >> to understand the percent distribution of data
langs = ["c", "python", "Java"]
students = [20, 100, 40]

plt.pie(students, labels = langs, autopct = '%1.1f%%')
plt.title("pie chart")
plt.show()
```

```python
# focus on each segment through separation

explode = (0.0, 0.1,0.1)
langs = ["c", "python", "Java"]
students = [20, 100, 40]

plt.pie(students, labels = langs, autopct = '%1.1f%%', explode = explode)
plt.title("pie chart")
plt.show()
```

```python
#shadow feature

explode = (0.0, 0.1,0.1)
langs = ["c", "python", "Java"]
students = [20, 100, 40]

plt.pie(students, labels = langs, autopct = '%1.1f%%', explode = explode, shadow = True)
plt.title("pie chart")
plt.show()
```

```python
years = [2021,2018,2022,2024,2077]
category1 = [20,25,34,12,42]
category2 = [19,23,45,35,12]
category3 = [2,5,7,3,9]

plt.stackplot(years, category1, category2, category3, labels = ['category1', 'category2', 'category3'], colors = ['skyblue','lightgreen','orange'])
plt.xlabel("years")
plt.ylabel("values")
plt.title("Area plot")
plt.legend()
plt.show()
```
