```python
# numerical data plotting
# relplot has options of plotting both categorical and numerical data
# relplot >> relations plot
# relplot >> relationship btw two variables
df1 = sns.load_dataset("tips")
```

```python
df1
```

```python
# how tip is varying w.r.t total_bill
sns.relplot(x = "total_bill", y = "tip", data = df1) # relationship btw two variables, by default it gives scatter plot
```

```python
sns.relplot(x = "total_bill", y = "tip", data = df1, hue = "smoker")
```

```python
sns.relplot(x = "total_bill", y = "tip", data = df1, hue = "size")
```

```python
sns.relplot(x = "total_bill", y = "tip", data = df1, style = "size")
```
